```python
from sklearn import datasets
```

```python
%matplotlib inline
```

```python
circles = datasets.make_circles()
```

```python
print(f"featires: {circles[0][:10]}")
print(f"target: {circles[1][:10]}")
```

```python
from matplotlib.colors import ListedColormap
```

```python
colors = ListedColormap(['red', 'yellow'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(list(map(lambda x: x[0], circles[0])), list(map(lambda x: x[1], circles[0])), c=circles[1], cmap=colors)
```

```python
def plot_2d(data, colors):
    plt.figure(figsize=(8, 8))
    plt.scatter(list(map(lambda x: x[0], data[0])), list(map(lambda x: x[1], data[0])), c=data[1], cmap=colors)
```

```python
noisy_circles = datasets.make_circles(noise=0.05)
```

```python
plot_2d(noisy_circles, colors)
```

```python
simple_classification_problem = datasets.make_classification(n_features=2, n_informative=1, n_redundant=1, 
                                                             n_clusters_per_class=1, random_state=1)
```

```python
plot_2d(simple_classification_problem, colors)
```

```python
classification_problem = datasets.make_classification(n_features=4, n_informative=3, n_redundant=1, n_classes=4,
                                                      n_clusters_per_class=1, random_state=1)
```

```python
plot_2d(classification_problem, ListedColormap(['red', 'green', 'black']))
```

```python
import pandas as pd
```

```python
iris_frame = pd.DataFrame(iris.data)
iris_frame.columns = iris.feature_names
iris_frame['target'] = iris.target
```

```python
iris.target_names
```

```python
iris_frame.target = iris_frame.target.apply(lambda x: iris.target_names[x])
```

```python
iris_frame
```

```python
iris_frame[iris_frame.target == 'virginica'].hist('petal width (cm)')
```

```python
plt.figure(figsize=(20, 24))

plot_index = 0
for feature_name in iris['feature_names']:
    for target_name in iris['target_names']:
        plot_index += 1
        plt.subplot(4, 3, plot_index)
        plt.hist(iris_frame[iris_frame.target == target_name][feature_name])
        plt.title(target_name)
        plt.xlabel('cm')
        plt.ylabel(feature_name[::-1])
```
